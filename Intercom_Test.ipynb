```python
#Read the libraries and load the data
import numpy as np, pandas as pd,json
from math import radians
from numpy import cos, sin, arcsin, sqrt
data=[]
for line in open('interc.json','r'):
    data.append(json.loads(line))
print(data)
```

```python
#Convert data into a dataframe
dataframe=pd.DataFrame.from_dict(data)
print(dataframe)
```

```python
#Convert Strings into Floats add new columns for office location
dataframe['office_lattitude']=53.339428
dataframe['office_longitude'] = -6.257664
dataframe['latitude']=dataframe['latitude'].astype(float)
dataframe['longitude']=dataframe['longitude'].astype(float)
dataframe['office_lattitude']=dataframe['office_lattitude'].astype(float)
dataframe['office_longitude']=dataframe['office_longitude'].astype(float)
```

```python
#Checking to see if the values are of the same datatype
print(dataframe.dtypes)
```

```python
#Define the Great Circle Distance function
def great_cirle(row):
    lon1 = -6.257664
    lat1 = 53.339428
    lon2 = row['longitude']
    lat2 = row['latitude']
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * arcsin(sqrt(a)) 
    km = 6367 * c
    return km
```

```python
#add a new column to the dataframe with the distance between office and the user
dataframe['distance_km'] = dataframe.apply(lambda row: great_cirle(row), axis=1)
```

```python
dataframe.drop(columns=['distance'])
```

```python
#Find all users with distance less or equal to 100 km
a = dataframe[['user_id','name']].loc[dataframe['distance_km'] <= 100].sort_values(by=['user_id'])
print(a)
```

```python
a.to_csv('output.txt', index=False, sep=' ', header=True)
```
